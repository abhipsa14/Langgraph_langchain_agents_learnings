## How to split by character-character text splitter?
This is the simplest method. This split based on a given character sequence, which defaults to "\n\n". Chunk length is measured by number of characters.
1. How the text is split: by single character separator
2. How the chunk size is measured: by number of characters

```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader("speech.txt")
docs=loader.load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter
textsplitter=CharacterTextSplitter(separator="\n\n",chunk_size=100,chunk_overlap=20)
textsplitter.split_documents(docs)
```

```python
data=""
with open("speech.txt") as f:
    data=f.read()
textsplitter=CharacterTextSplitter(chunk_size=100,chunk_overlap=20)
text=textsplitter.create_documents([data])
print(text[0])
print(text[1])
```
